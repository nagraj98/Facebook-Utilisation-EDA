# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

**Handling redundant data**
 
- The userid has no use for analysis, its just an identifier.
- We will remove the dob columns as the information is already captured in age column


**Handling duplicate data**

- For identifying duplicate data, it is better to have all columns except userid.
- So after handling duplicate data, we will drop the dob columns.

```python
data_new = data.drop(['userid'],axis=1)
data_new.head()
```

```python
data_new[data_new.duplicated()]
```


**Handling duplicate data**

- As seen above, there are 8 duplicate records. 
- In a dataset like this, it can be possible to have people with same birthdate,gender to have joined a platform on the same day and have same number of friends.
- There can be coincidences, so it is safe to keep these records.
- But as these are only 8 records (0.008 %), we can drop these using the drop_duplicates method of pandas.

```python
data_new.drop_duplicates(inplace=True)
```

- Before dropping the dob columns, we just check if there is any inconsistency related to these columns.
- We check if there are any records where age and dob_year are inconstistent with the fact that this data is of 2013.

- As seen below, there is no such inconsistent record.

```python
data_new.loc[(data_new.age + data_new.dob_year)!=2013]
```

```python
# dropping the dob columns after handling duplicates and checking for inconsistency.
data_new.drop(['dob_day','dob_month','dob_year'],axis=1,inplace=True)
```

**Handling missing data**

The records with missing values in the tenure and gender column can be dropped as they form only 0.175% of the data.
This can be handled for each column as given in the below code snippet :

1. nullrows = data_new.index[(data_new.tenure.isnull()==True)|(data_new.gender.isnull()==True)]
2. data_new.drop(nullrows)

But as all the missing values are NaN, we can use the pandas dropna function.

```python
data_new.dropna(inplace=True)
```

**Handling outliers**

- As can be seen from the below histogram, there is a bump after an age of 100 which is odd. 
- We will handle these outliers by replacing these ages with the median age.

```python
sns.histplot(data_new.age)
```

- We are replacing the outlier age with median age.
- median() returns a float, so it needs to be converted to int otherwise the dtype of whole age column will be set to float.

```python
iqr = data_new.age.quantile(0.75)-data_new.age.quantile(0.25)
threshold = data_new.age.quantile(0.75) + 1.5*iqr
median = int(data_new.age.median())
data_new.loc[data_new.age>threshold,"age"]=median
```

```python
sns.histplot(data_new.age)
```

```python
'''
Handling the inconsistent data:

To determine inconsistency, we use some questions :

'''

# 1. Are there any records where the total likes are not equal to sum of mobile_likes and www_likes.

data_inc1 = data_new.loc[data_new.likes != (data_new.mobile_likes + data_new.www_likes)]
data_inc1['mobile_&_www_likes'] = data_inc1.mobile_likes + data_inc1.www_likes
print("\nRecords where the total likes are not equal to sum of mobile_likes and www_likes : \n")
print(data_inc1[['likes','mobile_&_www_likes']])

# 2. Are there any records where the total likes recieved are not equal to sum of mobile_likes_received and www_likes_received

data_inc2 = data_new.loc[data_new.likes_received != (data_new.mobile_likes_received + data_new.www_likes_received)]
data_inc2['mobile_&_www_likes_received'] = data_inc2.mobile_likes_received + data_inc2.www_likes_received
print("\n\nRecords where the total likes_received are not equal to sum of mobile_likes_received and www_likes_received : \n")
print(data_inc2[['likes_received','mobile_&_www_likes_received']])

# 3. As friendships initiated are supposed to be a subset of freind_count, check if there are any records with friendships_initiated more than the freind_count.

data_inc3 = data_new.loc[data_new.friend_count<data_new.friendships_initiated]
print("\n\nRecords where friendships_initiated are more than friend_count :\n")
print(data_inc3)
```

- As seen above , for question #1 and #2 we have 4 records each.
- These have a mere 1 or 2 count inconsistency which can be easily handled and it wont be needed to drop these records

- For question #3 there are no records,so no handling required.

```python
data_new.loc[(data_new.likes != (data_new.mobile_likes + data_new.www_likes)),"likes"] = data_new.mobile_likes + data_new.www_likes

data_new.loc[(data_new.likes_received != (data_new.mobile_likes_received + data_new.www_likes_received)),"likes_received"] = data_new.mobile_likes_received + data_new.www_likes_received
```

#### 1C Which age group gets more traction

Based on the age distribution that we saw earlier, we can divide the users into the following user groups -
1. Teens            (5 years ) : 13-17
2. Youth            (6 years ) : 17-23
3. Young adults     (10 years) : 23-33
4. Adults           (10 years) : 33-45
5. Old adults       (20 years) : 45-58
6. Senior citizens  (37 years) : above 58

```python
# Creating a categorical column "age-group" using the values from "age" column

def classify_ages(row):
    if row['age'] < 17:
        return 'teen'
    elif row['age'] < 23:
        return 'youth'
    elif row['age'] < 33:
        return 'young_adult'
    elif row['age'] < 45:
        return 'adult'
    elif row['age'] < 58:
        return 'old_adult'
    else:
        return 'senior_citizen'

data_new['age_group'] = data_new.apply(classify_ages,axis=1)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='age_group',data= data_new)

plt.title("Distribution of users based on age-group", size = 20, color = 'purple')
plt.xlabel('Age-group', size=15, color = 'purple')
plt.ylabel('Number of friends', size = 15, color = 'purple')
_ = plt.xticks(fontsize=13)
_ = plt.yticks(fontsize=12)
```

```python
# Comparing the pattern of friend_count based on agegroup
plt.figure(figsize=(14,6))
sns.boxplot(y='age_group', x= 'friend_count', showfliers=False, data = data_new)

plt.title("Pattern of friend-count based on age-group", size = 20, color = 'purple')
plt.ylabel('Age-group', size=15, color = 'purple')
plt.xlabel('Number of friends', size = 15, color = 'purple')
_ = plt.yticks(fontsize=14)
_ = plt.xticks(np.arange(0,810,50),fontsize=13)
```

```python
# Comparing the pattern of likes given based on agegroup
plt.figure(figsize=(14,6))
sns.boxplot(y='age_group', x= 'likes', showfliers=False, data = data_new)

plt.title("Pattern of likes given based on age-group", size = 20, color = 'purple')
plt.ylabel('Age-group', size=15, color = 'purple')
plt.xlabel('Number of likes given', size = 15, color = 'purple')
_ = plt.yticks(fontsize=14)
_ = plt.xticks(np.arange(0,300,15),fontsize=13)
```

### Q3 Which is more popular - mobile application or web application

For analysing the popularity of the two modes of access, we will check which was used more for the activity of liking a post.
We will make a new column called likes-mode which will have the following categories :
1. no_activity - the total number of likes given is 0
2. mobile - the likes given through the mobile are more than 65 percent of total likes.
3. www - the likes given through web appplication are more than 65 percent of the total likes
4. neutral - the likes given by either of the methods is not more than 65% of the total likes.


```python
# Making new categorical column 'likes-mode'

def classify_likes_mode(row):
    mob = row['mobile_likes']
    www = row['www_likes']
    if (mob+www != 0):
        mob_percent = (mob*100)/(mob+www)
        if mob_percent > 65:
            return 'mobile'
        elif mob_percent < 35:
            return 'www'
        else:
            return 'neutral'
    return 'no_activity'

data_new['likes_mode'] = data_new.apply(classify_likes_mode,axis=1)
```

```python
plt.figure(figsize=(10,6))

sns.set_palette("Dark2")
sns.countplot(data_new.likes_mode)

plt.title("User distribution based on mode of access", size = 20, color='maroon')
plt.ylabel('Number of users', size=15, color='maroon')
plt.xlabel('Mode used for liking', size = 15, color='maroon')
_ = plt.yticks(fontsize=12)
_ = plt.xticks(fontsize=13)
```

```python
plt.figure(figsize=(10,6))

sns.set_palette(sns.color_palette(["#af8dc3","#7fbf7b"]))
sns.countplot(x='likes_mode' , hue= 'gender', data= data_new)

plt.title("Gender distribution based on mode of access", size = 20, color='olive')
plt.ylabel('Number of users', size=15, color='olive')
plt.xlabel('Mode used for liking', size = 15, color='olive')
_ = plt.yticks(fontsize=12)
_ = plt.xticks(fontsize=13)
```
